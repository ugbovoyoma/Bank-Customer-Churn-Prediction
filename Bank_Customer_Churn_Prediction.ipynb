# **Main objective**

The primary objective of this project is to forecast whether a specific bank customer will continue to use them or stop using them (churn or not churn). In essence, churn prediction helps businesses move from a reactive approach (waiting for customers to leave) to a proactive one (taking steps to keep customers)




**Proactive Intervention:** Instead of reacting to customer loss, churn prediction enables businesses to take timely action.

**Key Actions May Include:**

**1. Targeted Retention:** Implement specific strategies to retain high-risk customers, such as special offers, improved service, personalized communication, and loyalty programs.

**2. Root Cause Analysis:** Analyze churn data to understand the underlying reasons for customer attrition and drive improvements in products, services, and customer experience.

**3. Strategic Decision-Making:** Inform critical business decisions, including resource allocation, pricing adjustments, and product development, to minimize churn and maximize profitability.

**4. Risk Assessment:** In certain sectors, like finance, use churn prediction to evaluate customer risk profiles.


.       

**Overall Impact:** Churn prediction empowers businesses to shift from a reactive stance to a proactive customer relationship management approach, fostering increased customer loyalty and long-term revenue stability.

# **Description of the data**


The bank customer churn dataset is a commonly used dataset for predicting customer churn in the banking industry. It contains information on bank customers who either left the bank or continue to be a customer. The dataset includes the following attributes:

**Customer ID:** A unique identifier for each customer

**Surname:** The customer's surname or last name

**Credit Score:** A numerical value representing the customer's credit score

**Geography:** The country where the customer resides (France, Spain or Germany)

**Gender:** The customer's gender (Male or Female)

**Age:** The customer's age.

**Tenure:** The number of years the customer has been with the bank

**Balance:** The customer's account balance

**NumOfProducts:** The number of bank products the customer uses (e.g., savings account, credit card)

**HasCrCard:** Whether the customer has a credit card (1 = yes, 0 = no)

**IsActiveMember:** Whether the customer is an active member (1 = yes, 0 = no)

**EstimatedSalary**: The estimated salary of the customer

**Exited:** Whether the customer has churned (1 = yes, 0 = no)

.

**Refrence:** https://www.kaggle.com/datasets/shubhammeshram579/bank-customer-churn-prediction

# **Import Data**

```python
#Import library for exploratory data analysis

import pandas as pd
```

```python
#Import Dataset

data = pd.read_csv('dataset.csv')
data.head()
```

```python
#Convert dataset to dataframe

dataframe = pd.DataFrame(data)
dataframe.head()
```

# **Exploratory Data Analysis**

```python
dataframe.describe()
```

```python
dataframe.info()
```

```python
dataframe['Exited'].value_counts()
```

```python
# I would be removing/deleting the surname column because i think it is not important for our model

dataframe= dataframe.drop(['Surname', 'RowNumber',	'CustomerId'], axis = 1)
dataframe.head()
```

```python
# Check if the data has missing values
dataframe.isnull().sum()
```

```python
# Delete the rows with missing entries
dataframe= dataframe.dropna()

dataframe.isnull().sum()
```

# **Feature Engineering**

```python
dataframe.columns
```

```python
# Convert categorical varibles to numerical(geography and gender column)

from sklearn.preprocessing import LabelEncoder

Categorical_Variables= dataframe[["Geography", "Gender"]]

encoder= LabelEncoder()

for col in Categorical_Variables:
  dataframe[col] = encoder.fit_transform(dataframe[col])
```

```python
# Scale numerical variables/features so that a particular feature with high capacity will not have an influence on others

from sklearn.preprocessing import StandardScaler

scaler= StandardScaler()

variables= ['CreditScore',	'Age',	'Tenure',	'Balance', 'NumOfProducts', 'EstimatedSalary']

dataframe[variables]= scaler.fit_transform(dataframe[variables])
```

```python
dataframe.head()
```

```python
# Split dataset into dependent and independent variables

X = dataframe.drop(['Exited'], axis= 1)
y= dataframe['Exited']
```

# **Model Building**

```python
# Split dataset into training and test set

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test= train_test_split(X,y, test_size=0.3, random_state= 42 )
```

```python
# Build a logistic regression model

from sklearn.linear_model import LogisticRegression

LR= LogisticRegression()


# Fit the already built model to the training dataset

LR.fit(X_train, y_train)
```

```python
# Predict

y_pred= LR.predict(X_test)
```

```python
# Logistic Regression Evaluation

from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix
```

```python
print('Accuracy Score:', accuracy_score(y_test, y_pred))
print('Precision Score:', precision_score(y_test, y_pred))
print('Recall Score:', recall_score(y_test, y_pred))
print('F1 Score:', f1_score(y_test, y_pred))
```

```python
# Build for KNearest Neighbors
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
```

```python
KN= KNeighborsClassifier()

KN.fit(X_train,y_train)
```

```python
y_pred2= KN.predict(X_test)

#Evaluate
print('Accuracy Score:', accuracy_score(y_test, y_pred2))
print('Precision Score:', precision_score(y_test, y_pred2))
print('Recall Score:', recall_score(y_test, y_pred2))
print('F1 Score:', f1_score(y_test, y_pred2))
```

```python
# Build for support vector machine
```

```python
SV= SVC()

SV.fit(X_train, y_train)
```

```python
y_pred3 = SV.predict(X_test)
```

```python
print('Accuracy Score:', accuracy_score(y_test, y_pred3))
print('Precision Score:', precision_score(y_test, y_pred3))
print('Recall Score:', recall_score(y_test, y_pred3))
print('F1 Score:', f1_score(y_test, y_pred3))
```

```python
#Build for Decision Tree

DT= DecisionTreeClassifier()

DT.fit(X_train, y_train)
```

```python
y_pred4= DT.predict(X_test)
```

```python
print('Accuracy Score:', accuracy_score(y_test, y_pred4))
print('Precision Score:', precision_score(y_test, y_pred4))
print('Recall Score:', recall_score(y_test, y_pred4))
print('F1 Score:', f1_score(y_test, y_pred4))
```

```python
# Build for Random Forest

RF= RandomForestClassifier()

RF.fit(X_train,y_train)
```

```python
y_pred5= RF.predict(X_test)
```

```python
print('Accuracy Score:', accuracy_score(y_test, y_pred5))
print('Precision Score:', precision_score(y_test, y_pred5))
print('Recall Score:', recall_score(y_test, y_pred5))
print('F1 Score:', f1_score(y_test, y_pred5))
```
